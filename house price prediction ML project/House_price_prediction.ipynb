```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Kolkata.csv')
```

```python
data.drop(columns=['Gymnasium'])
```

```python
data.dropna(inplace=True)
```

```python
print(data.columns)
```

```python
data['Location']=data['Location'].apply(lambda x:x.strip())
```

```python
data['Location'].value_counts()
```

```python
Location_stats = data.groupby('Location')['Location'].agg('count').sort_values(ascending=False)
```

```python
Locations_less_than_10_entries = Location_stats[Location_stats<=10]
```

```python
Locations_less_than_10_entries
```

```python
data['Location']=data['Location'].apply(lambda x:'other' if x in Locations_less_than_10_entries else x)
```

```python
data['Location'].value_counts()
```

```python
data
```

```python
data.Area.unique()
```

```python
data.dropna(inplace=True)
```

```python
data.describe()
```

```python
data
```

```python
data.drop(columns=['Gymnasium','SwimmingPool','CarParking','AC','LiftAvailable'],axis =1,inplace=True)
```

```python
data
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split

col_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ['Location']),
    remainder='passthrough'
)

lr = LinearRegression()
scaler = StandardScaler()

model = make_pipeline(col_trans, scaler, lr)
```

```python
col_trans=make_column_transformer((OneHotEncoder(sparse_output=False),['Location']),remainder='passthrough')
```

```python
data_input= data.drop(columns = ['Price'])
data_output=data['Price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_input, data_output, test_size=0.2, random_state=42)
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
input=pd.DataFrame([[1200,'Barasat',3,2]], columns =['Area','Location','No. of Bedrooms','No. of Bathrooms'])
```

```python
model.predict(input)
```
